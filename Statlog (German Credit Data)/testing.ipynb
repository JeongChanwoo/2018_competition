# Cross validation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
from sklearn.metrics import confusion_matrix
```

```python
kfold = KFold(n_splits=5 , random_state=22)
average_cv = []
accuracy = []
std_cv=[]

classifiers = ['Logistic_Reg', 'SGD_classfier', 'KNN', 'Naive_bayes', 'SVM','Decision_tree', 'Random_forest']
models = [LogisticRegression(), SGDClassifier(), KNeighborsClassifier(n_neighbors=10),GaussianNB(),
          svm.SVC(kernel='rbf'),DecisionTreeClassifier(),RandomForestClassifier(n_estimators=100)]

for model in models:
    
    cv_result = cross_val_score(model,x_data_stack, y_data, cv = kfold, scoring='accuracy')
    average_cv.append(cv_result.mean())
    std_cv.append(cv_result.std())
    accuracy.append(cv_result)

baseline_estimate_df = pd.DataFrame({'cv_mean' : average_cv,
                                                        'cv_std' : std_cv},index = classifiers)
baseline_estimate_df
```

```python
plt.figure(figsize=(12,6))
plt.grid()
plt.yticks(np.arange(0,1,0.05))
plt.ylim(0.40)
sns.boxplot(x=classifiers, y= accuracy)
```

```python
plt.figure(figsize=(10,5))
plt.title('Average cv accuracy')
sns.barplot(x=baseline_estimate_df['cv_mean'], y = classifiers, )
```
