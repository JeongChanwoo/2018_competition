```python
def column_unique(data):
    unique_dict={}
    count_dict={}
    for column in data.columns:
        unique_dict[column] =  data[column].unique()
        count_dict[column] =  data[column].value_counts()
    
    return unique_dict, count_dict
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import Binarizer
from sklearn.preprocessing import LabelBinarizer
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names]
```

```python
class MyLabelBinarizer(BaseEstimator, TransformerMixin):#LabelBinarizer Pipline 지원 안함  0.19버전 부터
    def __init__(self, *args, **kwargs): 
        self.encoder = LabelBinarizer(*args, **kwargs)
    def fit(self, X, y=None):
        self.encoder.fit(X)
        return self
    def transform(self, X, y=None):
        return self.encoder.transform(X)
```

```python
class Get_dummies(BaseEstimator, TransformerMixin):
    def __init__(self,attribute_names = categoric_column):
        self.attribute_names = attribute_names
        
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return pd.get_dummies(X[self.attribute_names])
```

```python
num_pipline = Pipeline([
    ('selector' , DataFrameSelector(numeric_column)),
    ('std_scaler', StandardScaler()),
])

ordinary_pipline = Pipeline([
    ('selector' , DataFrameSelector(ordinal_column)),
    ('mm_scaler' , MinMaxScaler()),
])

binary_pipline = Pipeline([
    ('selector' , DataFrameSelector(binary_column)),
#     ('L_binarizer', MyLabelBinarizer()),
])

cat_pipline = Pipeline([
    ('selector', DataFrameSelector(categoric_column)),
    ('get_dummies', Get_dummies())
])
```

```python
from sklearn.pipeline import FeatureUnion
```

```python
full_pipline = FeatureUnion(transformer_list=[
    ("num_pipline", num_pipline),
    ("ordinary_pipline", ordinary_pipline),
    ("binary_pipline", binary_pipline),
    ("cat_pipline", cat_pipline),
])
```

```python
train_prepared = full_pipline.fit_transform(train_set)
```
