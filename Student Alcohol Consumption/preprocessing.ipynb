# Preprocessing

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import Binarizer
from sklearn.preprocessing import LabelBinarizer
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names]
```

```python
class MyLabelBinarizer(BaseEstimator, TransformerMixin):#LabelBinarizer Pipline 지원 안함  0.19버전 부터
    def __init__(self,  attribute_names): 
        self.encoder_dict = {}
#         self.encoder = LabelBinarizer()
        self.attribute_names = attribute_names
        
        
    def fit(self, X, y=None):
        for attribute in self.attribute_names:
            lb = LabelBinarizer()
            self.encoder_dict[attribute] = lb.fit(X.loc[:, attribute])
            
#             self.encoder.fit(X)
        return self
    def transform(self, X, y=None):
        X = X[self.attribute_names]
        for col_index,attribute in enumerate(self.attribute_names):
            X.loc[:,attribute] = self.encoder_dict[attribute].transform(X.loc[:,attribute])
#             X[attribute] = self.encoder_dict[attribute].transform(X[attribute])
        
        return X
#         return self.encoder.transform(X)
```

```python
class Get_dummies(BaseEstimator, TransformerMixin):
    def __init__(self,attribute_names = categoric_column):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return pd.get_dummies(X[self.attribute_names])
```

```python
class Trans_dataframe(BaseEstimator, TransformerMixin):
    def __init__(self,attribute_names ):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return pd.DataFrame(X, columns=self.attribute_names)
```

```python
num_pipline = Pipeline([
    ('selector' , DataFrameSelector(numeric_column)),
      ('std_scaler', StandardScaler()),
    ("trand_df" , Trans_dataframe(numeric_column))
])

ordinary_pipline = Pipeline([
    ('selector' , DataFrameSelector(ordinal_column)),
    ('mm_scaler' , MinMaxScaler()),
    ("trand_df" , Trans_dataframe(ordinal_column))
])

binary_pipline = Pipeline([
    ('selector' , DataFrameSelector(binary_column)),
    ('L_binarizer', MyLabelBinarizer(binary_column)),
    ("trand_df" , Trans_dataframe(binary_column))
])

cat_pipline = Pipeline([
    ('selector', DataFrameSelector(categoric_column)),
    ('get_dummies', Get_dummies())
])
```

```python
def preprocessing_data(data=None, index=None, kind = None):
    if str(kind)=='train':
        
        
        train_prepared_id = pd.DataFrame(train_id)
        train_prepared_num = num_pipline.fit_transform(data)#.reindex(index)
        train_prepared_ord = ordinary_pipline.fit_transform(data)#.reindex(index)
        train_prepared_bin = binary_pipline.fit_transform(data)#.reindex(index)
        train_prepared_cat = cat_pipline.fit_transform(data)
        
        train_prepared = pd.concat([train_prepared_id,train_prepared_num,train_prepared_ord, 
                                    train_prepared_bin,train_prepared_cat], axis=1)
        
        return train_prepared
        
        
    elif str(kind) == 'test':
        
        test_prepared_id = pd.DataFrame(test_id)
        test_prepared_num = num_pipline.transform(data)#.reindex(index)
        test_prepared_ord = ordinary_pipline.transform(data)#.reindex(index)
        test_prepared_bin = binary_pipline.transform(data)#.reindex(index)
        test_prepared_cat = cat_pipline.transform(data)
        
        test_prepared = pd.concat([test_prepared_id,test_prepared_num,test_prepared_ord, 
                                   test_prepared_bin,test_prepared_cat], axis=1)
        
        return test_prepared
        
    elif str(kind)=='total':
        total_prepared_id = pd.DataFrame(total_id)
        total_prepared_num = num_pipline.fit_transform(data)#.reindex(index)
        total_prepared_ord = ordinary_pipline.fit_transform(data)#.reindex(index)
        total_prepared_bin = binary_pipline.fit_transform(data)#.reindex(index)
        total_prepared_cat = cat_pipline.fit_transform(data)
        
        total_prepared = pd.concat([total_prepared_id,total_prepared_num,total_prepared_ord, 
                                   total_prepared_bin,total_prepared_cat], axis=1)
        
        return total_prepared
```

```python
train_prepared = preprocessing_data(train_set, index=train_index, kind='train')
```

```python
test_prapred = preprocessing_data(test_set, index=test_index, kind='test')
```
